```python
import requests
import matplotlib.pyplot as plt
import time


latitude = []
longitude = []
timestamp = []

def iss_location():
    response = requests.get('http://api.open-notify.org/iss-now.json')
    data = response.json()
    timestamp.append(data['timestamp'])
    latitude.append(float(data['iss_position']['latitude']))
    longitude.append(float(data['iss_position']['longitude']))


numrequests = 102
delay = 10

for _ in range(numrequests):
    iss_location()
    time.sleep(delay)

     
plt.plot(longitude, latitude)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('ISS Path')
plt.show()
```
